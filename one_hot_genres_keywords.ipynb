```python
import numpy as np
import pandas as pd
from matplotlib import  pyplot as plt
```

```python
movie_data=pd.read_csv('./movie_metadata.csv') # read file
```

```python
genres=movie_data['genres'] # extract genres
keywords=movie_data['plot_keywords'] # extract keywords
```

```python
genres_data=genres.str.split('|',expand=True) # split into multiple columns
keywords_data=keywords.str.split('|',expand=True) # same for keywords
genres_data
```

```python
genres_dummies=pd.get_dummies(genres_data,prefix='',prefix_sep='') # one-hot enocde, but don't make new names for duplicates
keywords_dummies=pd.get_dummies(keywords_data,prefix='',prefix_sep='') # same for keywords 
```

```python
genres_grouped=genres_dummies.groupby(level=0, axis=1) #group columns with same name
genres_consolidated=genres_grouped.any() #check if any column with the same name is True. This is ultimately what we want.

keywords_grouped=keywords_dummies.groupby(level=0, axis=1)
keywords_consolidated=keywords_grouped.any()
genres_consolidated
```

```python
genres_consolidated=genres_consolidated.astype(int)
keywords_consolidated=keywords_consolidated.astype(int)
```

```python
genres_filtered=genres_consolidated.loc[:,genres_consolidated.sum()>10]
```

```python
keywords_filtered=keywords_consolidated.loc[:,keywords_consolidated.sum()>5]
```

```python
new_movie_data=pd.concat([movie_data.drop(['genres','plot_keywords'],axis=1),genres_filtered,keywords_filtered],axis=1)
new_movie_data.columns
```

```python
new_movie_data.to_csv('./one_hot_movie_metadata.csv',index=False)
```
